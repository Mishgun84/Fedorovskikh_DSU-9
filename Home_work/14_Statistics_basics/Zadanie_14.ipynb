### Выбросы

```python
# Выбросы опять же имеет смысл считатть только для количественных данных
```

```python
def outliers (column):
    """
    Функция для выявления выбросов в полученном столбце column через межквартльный размах. 
    Значения NaN также за пределами размаха, но их мы считать выбросам не будем, т.к. это просто отсутствие информации.
    Поэтому строки с NAN отражать в результатах не будем
    """
#     column = column.dropna()
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = result_data[column.between(lower_bound, upper_bound, inclusive=True)]
    
    return pd.concat([result_data, remove_outliers]).drop_duplicates(keep=False).dropna()
```

```python
# Для столбца rectal temperature выбросами будут значения меньше 36.75 и больше 39.55
outliers(result_data['rectal temperature'])
```

```python
# Для столбца pulse выбросами будут значения больше 148. Нижняя расчетная граница диапазона -12 для пульса бессмысленна.
# В любом случае у живой лошади значение пульса должно быть больше нуля.
outliers(result_data['pulse'])
```

```python
# Для столбца respiratory rate выбросами будут значения больше 62.25. Аналогично пульсу нижняя расчетная граница не имеет смысла
outliers(result_data['respiratory rate'])
```
