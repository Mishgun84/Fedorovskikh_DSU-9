## Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
joined = ratings.merge(movies, on='movieId', how='left') # Объединяем файлы в одну таблицу, чтобы добавить названия фильмов
joined.head()
```

```python
rating5 = joined[ joined.rating == 5.0] # Оставляем только фильмы с рейтингом 5.0
rating5['title'].value_counts().head(1) # Считаем сколько раз встречается каждый фильм. На первой строчке фильм с наибольшим показателем
```

## Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
filtered_countries = data[ 
    ((data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia')) &
    ((data['category']==4) | (data['category']==12) | (data['category']==21)) &
    (data['year'] > 2005) & (data['year'] < 2010) & (data['quantity'] >= 0)]
print(f"Суммарное потребление стран Прибалтики: {filtered_countries['quantity'].sum()}")
```

```python
# второй вариант
data.loc[ 
    ((data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia')) &
    ((data['category']==4) | (data['category']==12) | (data['category']==21)) &
    (data['year'] > 2005) & (data['year'] < 2010) & (data['quantity'] >= 0)]['quantity'].sum()
```
